```python
import random
import pandas as pd
import numpy as np
```

```python
random.seed('АДВИ_2023')
```

```python
N = 1000
```

```python
indx_lst = [i for i in range(1,N+1)]

random.shuffle(indx_lst)

indx_lst
```

```python
groups = ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"]
groups_lst = random.choices(population = groups, k=N)

groups_lst
```

```python
uniform_lst = [random.randint(50, 100) for i in range(1,N+1)]
uniform_lst
```

```python
mu_1, s_1 = 0, 1
mu_2, s_2 = 0, 12
mu_3, s_3 = 50, 9
```

```python
gauss_1 = [random.gauss(mu_1, s_1) for i in range(1,N+1)]
gauss_2 = [random.gauss(mu_2, s_2) for i in range(1,N+1)]
gauss_3 = [random.gauss(mu_3, s_3) for i in range(1,N+1)]
```

```python
len(groups_lst) == len(uniform_lst) == len(gauss_2) == 1000
```

```python
d = {'groups': groups_lst, 'uniform': uniform_lst, 'gauss_1': gauss_1,'gauss_2': gauss_2,'gauss_3': gauss_3}
```

```python
df = pd.DataFrame(data = d,index = indx_lst)
```

```python
df.head()
```
